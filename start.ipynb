image reflection 

```python
import numpy as np
import cv2 as cv
img = cv.imread('pexels-cottonbro-9303998.jpg',0)
rows,cols = img.shape
M=np.float32([[1,0,0],[0,-1,rows],[0,0,1]])
reflected_image=cv.warpPerspective(img,M,(int(cols),int(rows)))
cv.imshow('Reflected Image',reflected_image)
cv.imwrite('reflected_image.jpg',reflected_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
img = cv.imread('pexels-cottonbro-9303998.jpg',0)
rows,cols = img.shape
M=np.float32([[1,0,0],[0,-1,rows],[0,0,1]])
img_rotatirtion=cv.warpAffine(img,cv.getRotationMatrix2D((cols/2,rows/2),30,0.6),(cols,rows))
cv.imshow('Rotated Image',img_rotatirtion)
cv.imwrite('rotated_image.jpg',img_rotatirtion)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 
import numpy as np
img=cv2.imread(r"dhoni_drive.jpeg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
fourier_shift=cv2.dft(np.float32(gray),flags=cv2.DFT_COMPLEX_OUTPUT)
fourier_shift=np.fft.fftshift(fourier_shift)
magnitude=20*np.log(cv2.magnitude(fourier_shift[:,:,0],fourier_shift[:,:,1]))
magnitude=cv2.normalize(magnitude,None,0,255,cv2.NORM_MINMAX,cv2.CV_8UC1)
cv2.imshow('Fourier Transform',magnitude)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 
image = cv2.imread('dhoni_drive.jpeg',0)

DFT=cv2.dft(np.float32(image),flags=cv2.DFT_COMPLEX_OUTPUT)
shift=np.fft.fftshift(DFT)
row,col=image.shape
center_row,center_col=row//2,col//2
mask=np.zeros((row,col,2),np.uint8)
mask[center_row-30:center_row+30,center_col-30:center_col+30]=1

fft_shift=shift*mask
fft_ishift=np.fft.ifftshift(fft_shift)
img_back=cv2.idft(fft_ishift)
imageThen=cv2.magnitude(img_back[:,:,0],img_back[:,:,1])


plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('Original Image')     
plt.subplot(1,2,2)
plt.imshow(imageThen,cmap='gray')
plt.title('Image after Low Pass Filter')
plt.show()
```

```python
import cv2
import numpy as np
def pixelval(pix,r1,s1,r2,s2):
    if(pix<r1):
        return 0
    elif(r1<=pix and pix<=r2):
        m=(s2-s1)/(r2-r1)
        c=s1-m*r1
        return int(m*pix+c)
    else:
        return 255
img=cv2.imread('dhoni_drive.jpeg',0)
r1=70
s1=0
r2=140
s2=255
pixelval_vec=np.vectorize(pixelval)
contrast_stretched_img=pixelval_vec(img,r1,s1,r2,s2).astype(np.uint8)
cv2.imshow('Original Image',img)
cv2.imshow('Contrast Stretched Image',contrast_stretched_img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
cv2.imwrite('contrast_stretched_image.jpg',contrast_stretched_img)
```
