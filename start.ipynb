image reflection 

```python
import numpy as np
import cv2 as cv
img = cv.imread('pexels-cottonbro-9303998.jpg',0)
rows,cols = img.shape
M=np.float32([[1,0,0],[0,-1,rows],[0,0,1]])
reflected_image=cv.warpPerspective(img,M,(int(cols),int(rows)))
cv.imshow('Reflected Image',reflected_image)
cv.imwrite('reflected_image.jpg',reflected_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

image rotatition 

```python
import numpy as np
import cv2 as cv
img = cv.imread('pexels-cottonbro-9303998.jpg',0)
rows,cols = img.shape
M=np.float32([[1,0,0],[0,-1,rows],[0,0,1]])
img_rotatirtion=cv.warpAffine(img,cv.getRotationMatrix2D((cols/2,rows/2),30,0.6),(cols,rows))
cv.imshow('Rotated Image',img_rotatirtion)
cv.imwrite('rotated_image.jpg',img_rotatirtion)
cv.waitKey(0)
cv.destroyAllWindows()
```

enlarging image 

```python
import numpy as np
import cv2 as cv
img = cv.imread('pexels-cottonbro-9303998.jpg',0)
rows,cols = img.shape
img_shrinked=cv.resize(img,(250,200),interpolation=cv.INTER_AREA)
cv.imshow('Shrunk Image',img_shrinked)
img_enlarged=cv.resize(img_shrinked,None,fx=1.5,fy=1.5,interpolation=cv.INTER_CUBIC)
cv.imshow('Enlarged Image',img_enlarged)
cv.waitKey(0)
cv.destroyAllWindows()
```

shaering X - Y axis 

```python
import numpy as np
import cv2 as cv
img = cv.imread('pexels-cottonbro-9303998.jpg',0)
rows,cols = img.shape

M=np.float32([[1,0,0],[0.5,1,0],[0,0,1]])
sheared_img=cv.warpPerspective(img,M,(int(cols*1.5),int(rows*1.5)))
cv.imshow('Sheared Image',sheared_img)
cv.imwrite('sheared_image.jpg',sheared_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 
import numpy as np
img=cv2.imread(r"dhoni_drive.jpeg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
fourier_shift=cv2.dft(np.float32(gray),flags=cv2.DFT_COMPLEX_OUTPUT)
fourier_shift=np.fft.fftshift(fourier_shift)
magnitude=20*np.log(cv2.magnitude(fourier_shift[:,:,0],fourier_shift[:,:,1]))
magnitude=cv2.normalize(magnitude,None,0,255,cv2.NORM_MINMAX,cv2.CV_8UC1)
cv2.imshow('Fourier Transform',magnitude)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
