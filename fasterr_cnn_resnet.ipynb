```python
import torch 
import torchvision 
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
from torch.utils.data import Dataset

class ToyDataset(Dataset):
    def __init__(self):
        self.images = []
        self.targets = []

        # Create 2 fake images (random tensors)
        for i in range(2):
            img = torch.rand(3, 224, 224)

            boxes = torch.tensor([[50, 60, 150, 180]], dtype=torch.float32)
            labels = torch.tensor([1])  # class id

            target = {
                "boxes": boxes,
                "labels": labels
            }

            self.images.append(img)
            self.targets.append(target)

    def __getitem__(self, idx):
        return self.images[idx], self.targets[idx]

    def __len__(self):
        return len(self.images)

dataset = ToyDataset()
```

```python
model = fasterrcnn_resnet50_fpn(pretrained=True)
num_classes=2 
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = torchvision.models.detection.faster_rcnn.FastRCNNPredictor(in_features, num_classes)    
model.train()
```

Training loop

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.005, momentum=0.9, weight_decay=0.0005)
for epoch in range(2):
    for images, targets in dataset:
        images = [images]
        targets = [targets]

        loss_dict = model(images, targets)
        losses = sum(loss for loss in loss_dict.values())

        optimizer.zero_grad()
        losses.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {losses.item():.4f}")
```

```python
model.eval()
image,_=dataset[0]
with torch.no_grad():
    prediction = model([image])
print(prediction)
```

```python
def show_prediction(image, prediction):
    img = image.permute(1, 2, 0).numpy()
    plt.imshow(img)
    for box in prediction[0]['boxes']:
        x1, y1, x2, y2 = box
        rect = plt.Rectangle((x1, y1), x2-x1, y2-y1, fill=False, color='red')
        plt.gca().add_patch(rect)
    plt.show()
show_prediction(image, prediction)
```
