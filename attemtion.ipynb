```python
import math 
def matmul(A,B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result
```

```python
def transpose(M):
    return list(map(list, zip(*M)))
    
```

```python
def  softmax(matrix):
    result=[]
    for row in matrix:
        exp_row=[math.exp(x) for x in row]
        sum_exp_row=sum(exp_row)
        result.append([x/sum_exp_row for x in exp_row])
    return result 
```

```python
def scale(matrix,d):
    return [[x/math.sqrt(d) for x in row] for row in matrix]
```

#example with input and output 

```python
x=[[1,0,1],[0,1,0],[1,0,1]]
wq=[[1,0,1],[0,1,0],[1,0,1]]
wk=[[1,0,1],[0,1,0],[1,0,1]]
wv=[[1,0,1],[0,1,0],[1,0,1]]
```

```python
q= matmul(x,wq)
k= matmul(x,wk)
v= matmul(x,wv)
```

```python
kt=transpose(k)
scores= matmul(q,kt)
```

```python
scaled_scores=scale(scores,d=2)
attention_weights=softmax(scaled_scores)
output= matmul(attention_weights,v)
print("Attention Weights:")
for row in attention_weights:
    print(row)
print('\nFinal Output:')
for row in output:
    print(row)
```

```python
def zeros(r,c):
    return [[0.0]*c for _ in range(r)]
def randn(r,c,scale=0.02):
    return [[random.random()*2-1 for _ in range(c)] for _ in range(r)]
def transpose(A):
    return list(map(list, zip(*A)))
def matmul(A,B):
    n,m=len(A),len(A[0])
    p=len(B[0])
    out=zeros(n,p)
    for i in range(n):
        Ai=A[i]
        for k in range(m):
            a=Ai[k]
            Bk=B[k]
            for j in range(p):
                out[i][j]+=a*Bk[j]
    return out
```

```python
def add(A,B):
    return [[A[i][j]+B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
```

```python
def add_inplace(A,B,alpha=1.0):
    for i in range(len(A)):
        for j in range(len(A[0])):
            A[i][j]+=alpha*B[i][j]
```

```python
def mul_scalar(A,scalar):
    return [[x*scalar for x in row] for row in A]
```

```python
def relu(matrix):
    return [[max(0,x) for x in row] for row in matrix]
```

```python
def relu_backward(dout,cache):
    x=cache
    dx=[[dout[i][j] if x[i][j]>0 else 0 for j in range(len(x[0]))] for i in range(len(x))]
    for i in range(len(dx)):
        for j in range(len(dx[0])):
            if x[i][j]<=0:
                dx[i][j]=dout[i][j] if x[i][j]>0 else 0
    return dx
```
